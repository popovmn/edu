List Statistics
---------------

You will be given a number n. On the next n lines you will receive
integers. You have to create and print two lists:

-   One with all the positives (including 0)

-   One with all the negatives

Finally print the following message**: "Count of positives:
{count\_positives}. Sum of negatives: {sum\_of\_negatives}"**

### Hints

We start by reading the number n:



Then, we create **a loop**, we read the **current number** and check if
it **is greater than zero or not**



-   If it is, we add it to the positives list

-   If it is not, we add it to the negatives list

Then we print the three lines:



-   To get the count of the positives we can use the **len** function

-   To get the sum of the negatives we can use the **sum** function
